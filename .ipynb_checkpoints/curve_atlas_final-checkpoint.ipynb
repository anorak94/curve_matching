## cost functions
$\displaystyle k={\frac {x'y''-y'x''}{\left({x'}^{2}+{y'}^{2}\right)^{\frac {3}{2}}}}$
formula for the curvature of a plane curve where x and y are parameterized by t   
cost of matching is = Z ξ2 ξ1 | cos(ψ) − sin(ψ)|+ R|κ(h) cos(ψ) − κ¯(h¯)sin(ψ)|dξ


## solve dp for curve matching 

```python
def get_psi(ds1, ds2):
    """
    given the two ds segments computes the cosine and sine of the angle between the matching 
    """
    hyp = (ds1**2 + ds2**2)**0.5
    cos_psi = ds1 / hyp
    sin_psi = ds2 / hyp
    
    return cos_psi, sin_psi
    
def get_delta_s(p1, p2):
    """
    computes the ds between two points on a curve 
    """
    x1, y1 = p1
    x2, y2 = p2
    ds = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return ds

def get_kappa_and_derivatives(a):
    """
    returns the curvature and gradients of a curve segment 
    """
    dx_dt = np.gradient(a[:, 0])
    dy_dt = np.gradient(a[:, 1])
    d2x_dt2 = np.gradient(dx_dt)
    d2y_dt2 = np.gradient(dy_dt)
    curvature = np.abs(d2x_dt2 * dy_dt - dx_dt * d2y_dt2) / (dx_dt * dx_dt + dy_dt * dy_dt)**1.5
    return curvature, [dx_dt, dy_dt, d2x_dt2, d2y_dt2] 


def get_segments_from_pt_list(p1):
    seg_list = []
    curv_list = []
    kappa , _ = get_kappa_and_derivatives(p1)
    for i in range(1, len(p1)):
        
        strt = p1[i-1]
        end  = p1[i]
        ds   = get_delta_s(strt, end)
        seg_list.append(ds)
        curv_list.append(0.5*(kappa[i] + kappa[i-1]))
    return seg_list, curv_list


def interpolate_curve_segment(points, npoints, graph = False):
    """
    interpolates the curve segment between a set of points 
    """
    
    distance = np.cumsum( np.sqrt(np.sum( np.diff(points, axis=0)**2, axis=1 )) )
    distance = np.insert(distance, 0, 0)/distance[-1]

    
    if len(points) < 4:
        kind = "linear"
    else:
        kind = "cubic"
    interpolator =  interp1d(distance, points, kind=kind, axis=0)
    

    # Computed the spline for the asked distances:
    alpha = np.linspace(0, 1, npoints)
    points_fitted = interpolator(alpha)

    # Graph:
    if graph:
        plt.plot(*points.T, 'ok', label='original points');
        plt.plot(*points_fitted.T, '-r', label='fitted spline k=3, s=.2');
        plt.axis('equal'); plt.legend(); plt.xlabel('x'); plt.ylabel('y');
    else:
        pass
    return points_fitted    

def get_unique_elements(p1):
    """
    returns only unique pts from the pt list 
    """
    pint = []
    pset = set()
    for pi in p1:
        if (pi[0], pi[1]) in pset:
            pass
        else:
            pint.append(pi)
            pset.add((pi[0], pi[1]))
    return pint

def get_interpolated_segments_and_kappas(str_tup, end_tup, pt_list1, pt_list2, npts):
    """
    given the end points of the two curve segments
    computes the segment list and the curvature of those segments 
    """
    
    i1, i2 = str_tup
    j1, j2 = end_tup
    p1_list = pt_list1[i1:i2+1]
    p2_list = pt_list2[j1:j2+1]
    
    if i1 == i2:
        seg1 = [0 for i in range(npts-1)]
        curv1 = [0 for i in range(npts-1)]
        pt2_interp = interpolate_curve_segment(p2_list, npts, False)
        seg2, curv2 = get_segments_from_pt_list(pt2_interp)
        
    elif j1 == j2:
        seg2 = [0 for i in range(npts-1)]
        curv2 = [0 for i in range(npts-1)]
        pt1_interp = interpolate_curve_segment(p1_list, npts, False)
        seg1, curv1 = get_segments_from_pt_list(pt1_interp)
        
    else:
        pt1_interp = interpolate_curve_segment(p1_list, npts, False)
        pt2_interp = interpolate_curve_segment(p2_list, npts, False)
        seg1, curv1 = get_segments_from_pt_list(pt1_interp)
        seg2, curv2 = get_segments_from_pt_list(pt2_interp)
    
    return seg1, curv1, seg2, curv2, pt1_interp, pt2_interp
    
def compute_cost_match(ds1_list, ds2_list, kappa1, kappa2, R):
    """
    ds1_list , ds2_list, kappa1, kappa2 are equal sized lists containing the matches
    in terms of segment length
    """
    ct = 0
    for (ds1, ds2, k1, k2) in zip(ds1_list, ds2_list, kappa1, kappa2):
        cp, sp = get_psi(ds1, ds2)
        ds = ((ds1**2 + ds2**2)**0.5)
        ct = ct + ds*(abs(cp - sp) + R*(abs(k1*cp - k2*sp)))
    return ct 


def get_tup_list(i, j):
    """
    get valid tup list from all possible tuples 
    """
    str_tup_list = [(i, i), (i-1, i), (i-1, i), (i-2, i), (i-3, i), (i-3, i), (i-2, i), (i-1, i), (i-1, i)]
    end_tup_list = [(j-1, j), (j, j), (j-1, j), (j-1, j), (j-1, j), (j-2, j), (j-3, j), (j-3, j), (j-2, j)]
    s1_list = []
    s2_list = []
    for (str_tup, end_tup) in zip(str_tup_list, end_tup_list):
        s1, s2 = str_tup
        u1, u2 = end_tup
        if s1<0 or s2<0 or u1<0 or u2<0:
            pass
        else:
            s1_list.append(str_tup)
            s2_list.append(end_tup)
    
    return s1_list, s2_list
        
```

```python
## pass in the index of the segments getting matched and it should return the cost of the matching 

## start_seg = i1
## end_seg = i2

## start_seg = j1
## end_seg = j2

## pt_start  = i1
## pt_end = i2+1

## pt_start = j1
## pt_end = j2
## if i1 == i2 or j1 == j2 => psi is either 90 degrees or 0 degrees
## else interpolate the lines and compute the integral of the cost 

## match (i, j) and (i-3, j-1) i.e. i,i-1, i-2, i-3 gets matched with j and j-1
## match (i, j) and (i-3, j-2) i.e. i,i-1, i-2, i-3 gets matched with j, j

dist = np.zeros((len(p1_int), len(p2_int)))
for i in range(dist.shape[0]):
    for j in range(dist.shape[1]):
        dist[i][j] = np.inf
dist[0][0] = 0.0


R = 1.0
predecessor =[[(-1, -1) for i in range(dist.shape[1])] for j in range(dist.shape[0])]
for i in tqdm(range(1,dist.shape[0])):
    
    for j in range(1, dist.shape[1]):
        ## doing total matches here 
        options = []
        
        
            
        str_tup_list, end_tup_list = get_tup_list(i, j)
        for (str_tup, end_tup) in zip(str_tup_list, end_tup_list):
            
            seg1, curv1, seg2, curv2 = get_interpolated_segments_and_kappas(str_tup, end_tup, p1_int, p2_int, npts)
            ct = compute_cost_match(seg1, seg2, curv1, curv2, R)
            options.append(dist[str_tup[0]][end_tup[0]] + ct)
        trace_index = np.argmin(options)
        
        (st1, st2), (et1, et2) = str_tup_list[trace_index], end_tup_list[trace_index]
        #print (st1, et1)
        predecessor[st2][et2] = (st1, et1)
        
        dist [i][j] = min(options)
        

```

```python
src1, src2 = dist.shape[0]-1, dist.shape[1]-1
c = 0
global_path = [(src1, src2)]
while((src1, src2) != (0, 0)):
    global_path.append(predecessor[src1][src2])
    src1, src2 = predecessor[src1][src2]
```
